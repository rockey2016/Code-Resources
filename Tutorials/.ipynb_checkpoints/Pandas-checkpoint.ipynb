```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Pandas数据类型
## Series
### 初始化

```python
#一维 Series 可以用一维列表初始化
s1 = pd.Series([1,3,5,np.nan,6])
s1

# 默认情况下，Series 的下标都是数字（可以使用额外参数指定），类型是统一的
```

## DataFrame

```python
# Return a fixed frequency DatetimeIndex
date1 = pd.date_range('20130101', periods=6)
date1

np.random.randn(6,4)
df1 = pd.DataFrame(np.random.randn(6,4),index=date1,columns=['A','B','C','D'])
df1
# 默认情况下，如果不指定 index 参数和 columns，那么他们的值将用从 0 开始的数字替代
# 字典的每个 key 代表一列，其 value 可以是各种能够转化为 Series 的对象。

# 与 Series 要求所有的类型都一致不同，DataFrame 值要求每一列数据的格式相同：
# loc 可以方便的使用 label 进行索引
df1.loc[date1[0]]
# 选择多行多列
df1.loc[:,['A','C']]

# 连续切片
df1.iloc[3:5,1:2]
# 索引不连续的部分
df1.iloc[[1,2,4],[0,2]]

print('bool index')
# 布尔型索引
# 所有 A 列大于 0 的行
df1.A>0
df1[df1.A>0]

# 只留下所有大于 0 的数值
df1[df1>0]

print("copy dataframe")
df2 = df1.copy()
df2
df2['E'] = ['one', 'one','two','three','four','three']
df2

# 使用isin 方法做 filter 过滤
ｄf2['E'].isin(['two','four'])
df2[df2['E'].isin(['two','four'])]

type(df2['E'])
```

```python
s2 = pd.Series([1,2,3,4,5,6], index=pd.date_range('20130102', periods=6))
s2

#字典一样，直接指定F列的值为s1，此时以df已有的index为标准将二者进行合并，s1中没有的index项设为NaN，多余的项舍去：
df1['F'] = s2
df1

#使用 at 或 iat 修改单个值
df1.at[date1[0],'F'] = 0
df1

df1.iat[0,4] = 0.1
df1

# 设定满足条件的数值
df1[df1>0] = -df1
df1

# 与R中的 apply 操作类似，接收一个函数，默认是对将函数作用到每一列上
df1.apply(np.cumsum)

df1.apply(lambda x:x.max() - x.min())
```

```python
np.random.randint(0, 7, size=10)
s3 = pd.Series(np.random.randint(0, 7, size=10))
s3
s3.value_counts()
s3.hist()
```

```python
# 当 Series或者DataFrame的某一列是字符串时，我们可以用 .str 对这个字符串数组进行字符串的基本操作
s4 = pd.Series(['A', 'B', 'C', 'Aaba', 'Baca', np.nan, 'CABA', 'dog', 'cat'])
s4.str.lower()
```

### 合并操作

```python
df2 = pd.DataFrame(np.random.randn(10, 4))

df2
df2[:2]
df2[4:5]
df2[7:]

pieces = [df2[:2], df2[4:5], df2[7:]]
pieces
pd.concat(pieces)

# merge 可以实现数据库中的 join 操作
df3 = pd.DataFrame({'key':['bar','cow'],'lvaule':[1,2]})
df4 = pd.DataFrame({'key':['bar','cow'],'lvaule':[3,4]})
df3
df4
pd.merge(df3,df4,on='key')

# 向 DataFrame 中添加行
df5 = pd.DataFrame(np.random.randn(8, 4), columns=['A','B','C','D'])
df5
# 将第三行的值添加到最后
s = df5.iloc[3]
df5.append(s, ignore_index=True)

df6 = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar',
                          'foo', 'bar', 'foo', 'foo'],
                   'B' : ['one', 'one', 'two', 'three',
                          'two', 'two', 'one', 'three'],
                   'C' : np.random.randn(8),
                   'D' : np.random.randn(8)})
df6
df6.groupby('A')
df6.groupby('A').sum()

#按照 A, B 的值进行分类
df6.groupby(['A','B']).sum()
```

```python
l1 = [['bar', 'bar', 'baz', 'baz',
                     'foo', 'foo', 'qux', 'qux'],
                    ['one', 'two', 'one', 'two',
                     'one', 'two', 'one', 'two']]
l1
# zip的内容要经过list之后才能显示出来
# zip相当与压缩  zip（*）相当于解压
tuples = list(zip(*l1))
index = pd.MultiIndex.from_tuples(tuples,names=['first','second'])
print(index)
df = pd.DataFrame(np.random.randn(8, 2), index=index, columns=['A', 'B'])
df

# stack方法将 columns 变成一个新的 index 部分：
df2 = df[:4]
stacked = df2.stack()

stacked
# 可以使用 unstack() 将最后一级 index 放回 column
```
