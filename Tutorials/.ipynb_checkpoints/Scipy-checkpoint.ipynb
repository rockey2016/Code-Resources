# 概率统计方法

```python
from numpy import array
heights = array([1.46, 1.79, 2.01, 1.75, 1.56, 1.69, 1.88, 1.76, 1.88, 1.78])
heights

heights.mean() #平均数
heights.std()  #标准差
heights.var()  #方差
```

在数据序列呈对称分布（正态分布）的状态下，其均值、中位数和众数重合。且在这三个数的两侧，其它所有的数据完全以对称的方式左右分布。

偏度>0，分布右偏，长尾在右，高峰在左，这似乎与一般认知不太一致。但其实我们可以发现偏度实际上是三阶标准中心矩，
而一个数据距离“中心”越远，对中心矩的计算影响越大。而当数据长尾在右，即有更多正偏的离群值，因此偏度>0

峰度Kurtosis，是研究数据分布陡峭或平滑的统计量，通过对峰度系数的测量，我们能够判定数据分布相对于正态分布而言是更陡峭还是平缓。

对于随机变量X，我们定峰度为其的四阶标准中心矩。
峰度其实也是一个相对于正态分布的对比量，正态分布的峰度系数为0，而均匀分布的峰度为-1.2，指数分布的峰度为6。

当峰度系数>0，从形态上看，它相比于正态分布要更陡峭或尾部更厚；而峰度系数<0,从形态山看，则它相比于正态分布更平缓或尾部更薄。
在实际环境当中，如果一个分部是厚尾的，这个分布往往比正态分布的尾部具有更大的“质量”，即含又更多的极端值。


```python
import scipy.stats.stats as st
st.tmean(heights)
st.tmean(heights,(1.7,2.0)) #小于1.7或大于2.0的数将被忽略
st.mode(heights) #众数及出现次数
st.skew(heights) #偏度
st.kurtosis(heights) #峰度
```

常见的[连续概率分布](https://zh.wikipedia.org/wiki/Category:%E8%BF%9E%E7%BB%AD%E5%88%86%E5%B8%83)有：

- 均匀分布

[离散概率分布](https://zh.wikipedia.org/wiki/Category:%E7%A6%BB%E6%95%A3%E5%88%86%E5%B8%83)：

- 伯努利分布
- 几何分布
- ...

这些都可以在 `scipy.stats` 中找到。

- 正态分布
它包含四类常用的函数：

    norm.cdf 返回对应的累计分布函数值  
    norm.pdf 返回对应的概率密度函数值   
    norm.rvs 产生指定参数的随机变量  
    norm.fit 返回给定数据下，各参数的最大似然估计（MLE）值  
从正态分布产生500个随机点：


```python
from scipy.stats import norm
x_norm = norm.rvs(size=100)
```

```python
%pylab inline
x_norm
h = hist(x_norm)
type(h)
h[0] #次数
h[1] #bin centers
```

```python
#density如果是``True``，则返回元组的第一个元素将是是规范化以形成概率密度的计数，即直方图下的面积（或积分）将总和为1。
#通过将计数除以数量来实现的观察时间乘以箱宽，而不是除以总数观察次数。 如果* stacked *也是``True``，那么总和直方图归一化为1
#h = hist(x_norm,density=True,bins=20)
#h[0]
#h[1]
h1 = hist(x_norm,density=True,bins=20,stacked=True)
h1[0]
h1[1]
```

```python
from scipy.stats import norm
x_mean, x_std = norm.fit(x_norm)

print('mean, ', x_mean)
print('x_std, ', x_std)
```

```python
#将真实的概率密度函数与直方图进行比较
h = hist(x_norm, normed=True, bins=20)

x = linspace(-2,3,30)
p = plot(x, norm.pdf(x), 'r-')
```

# 线性代数

numpy 和 scipy 中，负责进行线性代数部分计算的模块叫做 linalg  
一方面scipy.linalg 包含 numpy.linalg 中的所有函数，同时还包含了很多 numpy.linalg 中没有的函数。  
另一方面，scipy.linalg 能够保证这些函数使用 BLAS/LAPACK 加速，而 numpy.linalg 中这些加速是可选的。  
因此，在使用时，我们一般使用 scipy.linalg 而不是 numpy.linalg。  

```python
import numpy as np
import numpy.linalg
import scipy as sp
import scipy.linalg
import matplotlib.pyplot as plt
from scipy import linalg

%matplotlib inline
```

 ## 求解线性方程组

例如，下列方程组
$$
x + 3y + 5z  =  10 \\
2x + 5y + z  =  8  \\
2x + 3y + 8z =  3
$$
的解为：
$$
\left[\begin{array}{c} x\\ y\\ z\end{array}\right]=\left[\begin{array}{ccc} 1 & 3 & 5\\ 2 & 5 & 1\\ 2 & 3 & 8\end{array}\right]^{-1}\left[\begin{array}{c} 10\\ 8\\ 3\end{array}\right]=\frac{1}{25}\left[\begin{array}{c} -232\\ 129\\ 19\end{array}\right]=\left[\begin{array}{c} -9.28\\ 5.16\\ 0.76\end{array}\right]
$$
可以使用 `linalg.solve` 求解方程组，也可以先求逆再相乘，两者中 `solve` 比较快。

```python
import time

A = np.array([[1, 3, 5],
              [2, 5, 1],
              [2, 3, 8]])
b = np.array([10, 8, 3])

tic = time.time()
A_1 = scipy.linalg.inv(A)
X = A_1.dot(b)
X

X1 = scipy.linalg.solve(A,b)
print(repr(X1))
```
